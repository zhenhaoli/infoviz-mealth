```python
import pandas as pd
import numpy as np

df = pd.read_csv('sterbealter.csv')
df.head()
```

```python
# df_cleaned_decimals = df_selected.applymap(lambda x: str(x).replace(",", "."))
# df_cleaned_decimals    
```

```python
def nummer2mapid(nummer):
    return {
        '1': 12,
        '2': 9,
        '3': 11,
        '4': 21,
        '5': 14,
        '6': 0,
        '7': 27,
        '8': 8,
        '9': 22, #18
        '10': 6,
        '11': 10,
        '12': 20,
        '13': 13,
        '14': 15,
        '15': 1,
        '16': 16,
        '17': 25,
        '18': 24,
        '19': 26, #28, 17
        '20': 2,
        '21': 23, #19
        '22': 3,
        '23': 4,
        '24': 5,
        '25': 7,
        '#100': 0
    }[nummer]
```

```python
df['mapId'] = df['NUMMER'].apply(nummer2mapid)
df.head()
```

```python
df_selected = df[['mapId', 'NAME', 'JAHR', 'INDIKATOR_AUSPRAEGUNG', 'INDIKATOR_WERT', 'GLIEDERUNG']]
df_selected.head()
```

```python
missing22to18 = []
missing26to28 = []
missing26to17 = []
missing23to19 = []
for index, row in df_selected.iterrows():
    if row['mapId'] == 22:
        missing22to18.append(row)
    elif row['mapId'] == 26:
        missing26to28.append(row)
        missing26to17.append(row)
    elif row['mapId'] == 23:
        missing23to19.append(row)

df18 = pd.DataFrame(missing22to18)
df18['mapId'] = 18

df28 = pd.DataFrame(missing26to28)
df28['mapId'] = 28

df17 = pd.DataFrame(missing26to17)
df17['mapId'] = 17

df19 = pd.DataFrame(missing23to19)
df19['mapId'] = 19

df19.head()
```

```python
frames = [df_selected, df18, df28, df17, df19]
result = pd.concat(frames)
result.head()
```

```python
districts_names = df_selected.sort_values(by=['NAME'])['NAME'].unique()
districts_names
```

```python
result.to_csv('sterbealter_cleaned.csv', index=False, encoding='utf-8-sig', sep=';')
```
